```python
"""
1. Write a Python program that: 
a. Creates a numpy-array of integers of size NxMx3, with N and M chosen dynamically by the 
operator (using the input function) containing randomly generated numbers in the range [100, 
5000] 
b. Define a function that swaps even-numbered columns with odd-numbered columns and vice 
versa 
c. Define a function that exchanges even rows with odd rows and vice versa 
d. Given a random number X as input, all pixels with a value equal to X must be set to 0 
"""
```

```python
import numpy as np
import random
import math as mt
```

```python
# 1A

def arr_creation():
    N = int(input("Rows: "))
    M = int(input("Columns: "))
    arr = np.random.randint(100, 5001, size = (N, M, 3))
    return arr
```

```python
I = arr_creation()
print(I)
```

```python
# 1B

def swap_columns(I):
    I2 = I.copy()
    for i in range(0, I.shape[0]):
        for j in range(0, I.shape[1], 2):
            I2[i,j] = I[i,j+1]
            I2[i,j+1] = I[i,j]
    return I2

print("Swap columns \n", swap_columns(I))

# 1C

def swap_rows(I):
    I2 = I.copy()
    for i in range(0, I.shape[0], 2):
        for j in range(0, I.shape[1]):
            I2[i,j] = I[i+1, j]
            I2[i+1,j] = I[i,j]
    return I2

print("\nSwap rows \n", swap_rows(I))

# 1D

def pixel_0(I):
    I2 = I.copy()
    X = int(input("Number between 100 and 5000: "))
    if X < 100 or X > 5000:
        raise ValueError
    for i in range(0, I.shape[0]):
        for j in range(0, I.shape[1]):
            for k in range(0, I.shape[2]):
                if I[i][j][k] == X:
                    I2[i][j][k] = 0
    return I2

print("\nCheck for 0 \n", pixel_0(I))
```

```python
# 2A

def arr_creation2():
    N = int(input("Rows: "))
    M = int(input("Columns: "))
    arr = np.random.randint(0, 256, size = (N, M, 3))
    return arr
```

```python
# 2B

def main_to_sec(I):
    I2 = I.copy()
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            if i == j: # Main diagonal element
                I2[i, j] = I[i, I.shape[0] - 1 - i] # Secondary diagonal element
                I2[i, I.shape[0] - 1 - i] = I[i, j]
    return I2

print("Exchange of diagonals \n", main_to_sec(I))

# 2C

def sec_to_zero(I):
    I2 = I.copy()
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            if j == I.shape[0] - 1 - i:
                I2[i, j] = 0
    return I2

print("\nSecondary diagonal to 0 \n", sec_to_zero(I))

# 2D

def binarization(I):
    I2 = I.copy()
    S = int(input("Threshold: "))
    if S < 0 or S > 255:
        raise ValueError
    I2[I <= S] = 0
    I2[I > S] = 255
    return I2

print("\nBinarization \n", binarization(I))
```

```python
# 4B + 4C

def odd_coor(I):
    I2 = I.copy()
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            for k in range(I.shape[2]):
                if i % 2 == 0 and j % 2 == 0 and k % 2 != 0:
                    neighbors = []
                    if i > 0: # Sopra
                        neighbors.append(I[i-1][j][k])
                    if i < I.shape[0] - 1: # Sotto
                        neighbors.append(I[i+1][j][k])
                    if j > 0: # Sinistra
                        neighbors.append(I[i][j-1][k])
                    if j < I.shape[1] - 1: # Destra
                        neighbors.append(I[i][j+1][k])
                    if neighbors:
                        I2[i][j][k] = sum(neighbors) / len(neighbors) # Media
                        
                elif i % 2 != 0 and j % 2 != 0 and k % 2 == 0:
                    neighbors = []
                    if i > 0: # Sopra
                        neighbors.append(I[i-1][j][k])
                    if i < I.shape[0] - 1: # Sotto
                        neighbors.append(I[i+1][j][k])
                    if j > 0: # Sinistra
                        neighbors.append(I[i][j-1][k])
                    if j < I.shape[1] - 1: # Destra
                        neighbors.append(I[i][j+1][k])
                    if neighbors:
                        I2[i][j][k] = abs(neighbors[0] - neighbors[1]) # Differenza in valore assoluto
    return I2

I2 = odd_coor(I)
print(I2)
```

```python
# 5A 

def identity(I):
    I2 = I.copy()
    return I2

print("Identity \n", identity(I))

# 5B

def downscaling(I):
    scale_factor = int(input("Scale by: "))
    new_h = int(I.shape[0] / scale_factor)
    new_w = int(I.shape[1] / scale_factor)
    I2 = np.zeros((new_h, new_w, I.shape[2]), dtype = I.dtype)
    for i in range(new_h):
        for j in range(new_w):
            I2[i,j] = I[int(i*scale_factor), int(j*scale_factor)]
    return I2

def upscaling(I):
    scale_factor = int(input("Scale by: "))
    new_h = int(I.shape[0] * scale_factor)
    new_w = int(I.shape[1] * scale_factor)
    I2 = np.zeros((new_h, new_w, I.shape[2]), dtype = I.dtype)
    for i in range(new_h):
        for j in range(new_w):
            I2[i,j] = I[int(i / scale_factor), int(j / scale_factor)]
    return I2
    
print("\nDownscaling \n", downscaling(I))
print("\nUpscaling \n", upscaling(I))
```

```python
# 5C

def rotation(I):
    angle = int(input("Angle: "))
    theta = mt.radians(angle)
    cos_theta = mt.cos(theta)
    sin_theta = mt.sin(theta)
    cx = I.shape[0] // 2 # centro x
    cy = I.shape[1] // 2 # centro y
    I2 = np.zeros((I.shape[0], I.shape[1], I.shape[2]), dtype = I.dtype)
    for i in range(I2.shape[0]):
        for j in range(I2.shape[1]):
            x = i - cx
            y = j - cy
            new_x = int(round(x*cos_theta - y*sin_theta))
            new_y = int(round(x*sin_theta + y*cos_theta))
            # Replication
            new_x = int(round(new_x)) 
            new_y = int(round(new_y))
            if 0 <= new_x <= I2.shape[0] and 0 <= new_y <= I2.shape[1]:
                I2[i,j] = I[new_x, new_y]
    return I2

print("\nRotation \n", rotation(I))
```

```python
# 5D

def translation(I):
    dx = int(input("Move the x by: "))
    dy = int(input("Move the y by: "))
    I2 = np.zeros((I.shape[0], I.shape[1], I.shape[2]), dtype = I.dtype)
    for i in range(I2.shape[0]):
        for j in range(I2.shape[1]):
            x = i - dx
            y = j - dy
            # Replication
            new_x = int(round(x))
            new_y = int(round(y))
            if 0 <= new_x <= I2.shape[0] and 0 <= new_y <= I2.shape[1]:
                I2[i,j] = I[x,y]
    return I2

print("\nTranslation \n", translation(I))
```

```python
# 7A

def decimation_avg(I):
    I2 = np.zeros((I.shape[0] // 2, I.shape[1] // 2, I.shape[2] // 2), dtype = I.dtype)
    for i in range(I2.shape[0]):
        for j in range(I2.shape[1]):
            block = I[i*2:(i+1)*2, j*2:(j+1)*2]
            I2[i,j] = np.mean(block)
    return I2

print("Decimation average \n", decimation_avg(I))
```

```python
# 7B

def decimation_min(I):
    I2 = np.zeros((I.shape[0] // 2, I.shape[1] // 2, I.shape[2] // 2), dtype = I.dtype)
    for i in range(I2.shape[0]):
        for j in range(I2.shape[1]):
            block = I[i*2:(i+1)*2, j*2:(j+1)*2]
            I2[i,j] = np.min(block)
    return I2

print("Decimation minimum \n", decimation_min(I))
```

```python
# 7C

def decimation_max(I):
    I2 = np.zeros((I.shape[0] // 2, I.shape[1] // 2, I.shape[2] // 2), dtype = I.dtype)
    for i in range(I2.shape[0]):
        for j in range(I2.shape[1]):
            block = I[i*2:(i+1)*2, j*2:(j+1)*2]
            I2[i,j] = np.max(block)
    return I2

print("Decimation maximum \n", decimation_max(I))
```

```python
# 9A

def bit_plane(I):
    bit_planes = []
    for bit in range(8):
        for channel in range(I.shape[2]):
            bit_planes.append((I[:, :, channel] >> bit) & 1) # Estrazione del bit 'bit' del canale 'channel'
    bit_planes_arr = np.array(bit_planes)
    return bit_planes_arr

BP = bit_plane(I)
print(BP)
```

```python
# 9B

def swap_bitplanes(BP): 
    swapped_BP = BP.copy()
    for row in range(0, BP.shape[0], 2):
        swapped_BP[row] = BP[row + 1]
        swapped_BP[row + 1] = BP[row]
    return swapped_BP

SW_BP = swap_bitplanes(BP)
print(SW_BP)
```

```python
# 10A, 10B, 10C

def bayer_1(I): # PATTERN GR-BG
    P1 = np.zeros((I.shape), dtype = I.dtype)
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            if i % 2 == 0: # riga pari
                if j % 2 == 0: # riga pari, colonna pari = canale verde
                    P1[i, j, 1] = I[i, j, 1] 
                else: # riga pari, colonna dispari = canale rosso
                    P1[i, j, 0] = I[i, j, 0]
            else: # riga dispari
                if j % 2 == 0: # riga dispari, colonna pari = canale blu
                    P1[i, j, 2] = I[i, j, 2]
                else: # riga dispari, colonna dispari = canale verde
                    P1[i, j, 1] = I[i, j, 1]
    return P1

def bayer_2(I): # PATTERN GR-GB
    P2 = np.zeros((I.shape), dtype = I.dtype)
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            if i % 2 == 0: # riga pari
                if j % 2 == 0: # riga pari, colonna pari = canale verde
                    P2[i, j, 1] = I[i, j, 1]
                else: # riga pari, colonna dispari = canale rosso
                    P2[i, j, 0] = I[i, j, 0]
            else: # riga dispari
                if j % 2 == 0: # riga dispari, colonna pari = canale verde
                    P2[i, j, 1] = I[i, j, 1]
                else: # riga dispari, colonna dispari = canale blu
                    P2[i, j, 2] = I[i, j, 2]
    return P2

def bayer_3(I): # PATTERN GR-GB-RG-GB
    P3 = np.zeros((I.shape), dtype = I.dtype)
    for i in range(I.shape[0]):
        for j in range(I.shape[1]):
            if i % 4 == 0: # riga 1 (GR)
                if j % 2 == 0: # riga 1, colonna pari = canale verde
                    P3[i, j, 1] = I[i, j, 1]
                else: # riga 1, colonna dispari = canale rosso 
                    P3[i, j, 0] = I[i, j, 0]
            elif i % 4 == 1: # riga 2 (GB)
                if j % 2 == 0: # riga 2, colonna pari = canale verde
                    P3[i, j, 1] = I[i, j, 1]
                else: # riga 2, colonna dispari = canale blu
                    P3[i, j, 2] = I[i, j, 2]
            elif i % 4 == 2: # riga 3 (RG)
                if j % 2 == 0: # riga 3, colonna pari = canale rosso
                    P3[i, j, 0] = I[i, j, 0]
                else: # riga 3, colonna dispari = canale verde
                    P3[i, j, 1] = I[i, j, 1]
            else: # riga 4 (BG)
                if j % 2 == 0: # riga 4, colonna pari = canale verde
                    P3[i, j, 2] = I[i, j, 2]
                else: # riga 4, colonna dispari = canale blu
                    P3[i, j, 1] = I[i, j, 1]
    return P3

print("Pattern 1 \n", bayer_1(I))
print("\nPattern 2 \n", bayer_2(I))
print("\nPattern 3 \n", bayer_3(I))
```

```python
# 10D, 10E, 10F (replication)

# Per mantenerci nei bordi, verifichiamo sempre che:
# 0 <= i < I.shape[0]
# 0 <= j < I.shape[1]

def rep_bayer_1(P1):
    repP1 = P1.copy()
    for i in range(P1.shape[0]):
        for j in range(P1.shape[1]):
            for c in range(P1.shape[2]):
                if c == 0: # canale rosso
                    if j - 1 >= 0: 
                        repP1[i, j-1, c] = P1[i, j, c] # interpolazione canale rosso
                    if i + 1 < P1.shape[0] and j - 1 >= 0:
                        repP1[i+1, j-1, c] = P1[i, j, c]
                    if i + 1 < P1.shape[0]:
                        repP1[i+1, j, c] = P1[i, j, c] 
                elif c == 1 and i % 2 == 0: # canale verde, riga pari
                    if j + 1 < P1.shape[1]:
                        repP1[i, j+1, c] = P1[i, j, c] 
                elif c == 1 and i % 2 != 0: # canale verde, riga dispari
                    if j - 1 >= 0:
                        repP1[i, j-1, c] = P1[i, j, c]
                elif c == 2: # canale blu
                    if i - 1 >= 0:
                        repP1[i-1, j, c] = P1[i, j, c]
                    if i - 1 >= 0 and j + 1 < P1.shape[1]:
                        repP1[i-1, j+1, c] = P1[i, j, c]
                    if j + 1 < P1.shape[1]:
                        repP1[i, j+1, c] = P1[i, j, c]
    return repP1

def rep_bayer_2(P2):
    repP2 = P2.copy()
    for i in range(P2.shape[0]):
        for j in range(P2.shape[1]):
            for c in range(P2.shape[2]):
                if c == 0: # canale rosso
                    if j - 1 >= 0:
                        repP2[i, j-1, c] = P2[i, j, c]
                    if j + 1 < P2.shape[1]:
                        repP2[i, j+1, c] = P2[i, j, c]
                    if i + 1 < P2.shape[0]:
                        repP2[i+1, j, c] = P2[i, j, c]
                elif c == 1: # canale verde
                    if j + 1 < P2.shape[1]:
                        repP2[i, j+1, c] = P2[i, j, c]
                elif c == 2: # canale blu
                    if j - 1 >= 0:
                        repP2[i, j-1, c] = P2[i, j, c]
                    if j + 1 < P2.shape[1]:
                        repP2[i, j+1, c] = P2[i, j, c]
                    if i - 1 >= 0:
                        repP2[i - 1, j, c] = P2[i, j, c]
    return repP2

def rep_bayer_3(P3):
    repP3 = P3.copy()
    for i in range(P3.shape[0]):
        for j in range(P3.shape[1]):
            for c in range(P3.shape[2]):
                if c == 0 and j % 2 != 0: # canale rosso, colonne dispari
                    if j - 1 >= 0:
                        repP3[i, j-1, c] = P3[i, j, c]
                    if j + 1 < P3.shape[1]:
                        repP3[i, j+1, c] = P3[i, j, c]
                    if i + 1 < P3.shape[0]:
                        repP3[i+1, j, c] = P3[i, j, c]
                elif c == 0 and j % 2 == 0: # canale rosso, colonne pari
                    if j + 1 < P3.shape[1]:
                        repP3[i, j+1, c] = P3[i, j, c]
                    if i + 1 < P3.shape[0]:
                        repP3[i+1, j, c] = P3[i, j, c]
                elif c == 1:
                    if i % 4 == 0 or i % 4 == 1: # canale verde, righe 1 e 2
                        if j + 1 < P3.shape[1]:
                            repP3[i, j+1, c] = P3[i, j, c]
                    else: # canale verde, righe 3 e 4
                        if j - 1 >= 0:
                            repP3[i, j-1, c] = P3[i, j, c]
                        elif j + 1 < P3.shape[1]:
                            repP3[i, j+1, c] = P3[i, j, c]
                elif c == 2:
                    if j % 2 == 0: # canale blu, colonna pari
                        if i - 1 >= 0:
                            repP3[i-1, j, c] = P3[i, j, c]
                        elif i + 1 < P3.shape[0]:
                            repP3[i+1, j, c] = P3[i, j, c]
                        elif j + 1 < P3.shape[1]:
                            repP3[i, j+1, c] = P3[i, j, c]
                    else: # canale blu, colonna dispari
                        if i - 1 >= 0:
                            repP3[i-1, j, c] = P3[i, j, c]
                        elif i + 1 < P3.shape[0]:
                            repP3[i+1, j, c] = P3[i, j, c]
                        elif j - 1 >= 0:
                            repP3[i, j-1, c] = P3[i, j, c]
                        elif j + 1 < P3.shape[1]:
                            repP3[i, j+1, c] = P3[i, j, c]
    return repP3

                        
print("Replication Pattern 1\n", rep_bayer_1(P1))
print("\nReplication Pattern 2\n", rep_bayer_2(P2))
print("\nReplication Pattern 3\n", rep_bayer_3(P3))

repP1 = rep_bayer_1(P1)
repP2 = rep_bayer_2(P2)
repP3 = rep_bayer_3(P3)

# 10G

def metrics_rep(P1, repP1, P2, repP2, P3, repP3):
    MN_1 = P1.size
    MN_2 = P2.size
    MN_3 = P3.size
    
    MSE_1 = np.sum((repP1 - P1)**2) / MN_1
    if MSE_1 != 0:
        PSNR_1 = -10 * mt.log10(MSE_1 / np.max(P1))
    else:
        PSNR_1 = float('inf')
        
    MSE_2 = np.sum((repP2 - P2)**2) / MN_2
    if MSE_2 != 0:
        PSNR_2 = -10 * mt.log10(MSE_2 / np.max(P2))
    else:
        PSNR_2 = float('inf')
        
    MSE_3 = np.sum((repP3 - P3)**2) / MN_3
    if MSE_3 != 0:
        PSNR_3 = -10 * mt.log10(MSE_3 / np.max(P3))
    else:
        PSNR_3 = float('inf')
        
    return [(MSE_1, MSE_2, MSE_3), (PSNR_1, PSNR_2, PSNR_3)]

metrics = metrics_rep(P1, repP1, P2, repP2, P3, repP3)
print("\nMSE values\n", metrics[0])
print("\nPSNR values\n", metrics[1])
```
